```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
survey_df  = pd.read_csv('survey_results_public.csv')

survey_df.head()
```

```python
languages = survey_df['LanguageWorkedWith'].to_frame()
languages
```

```python
c = set()
for i in list(languages['LanguageWorkedWith'].str.split(';')):
    try:
        for j in i:
            c.add(j)
    except:
        pass
        
c
```

```python
for lang in c:
    try:
        languages[lang] = languages['LanguageWorkedWith'].str.contains(lang, na=False)
    except:
        languages[lang] = False
```
