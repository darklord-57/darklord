```python
import requests
from bs4 import BeautifulSoup

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
url = 'https://www.webfx.com/tools/emoji-cheat-sheet/'

response = requests.get(url)
if response.status_code != 200:
    print(f"Failed to get the webpage. Status code: {response.status_code}")
    exit()
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

main_content = soup.find('div', {'class': 'ecs_main-content'})
```

```python
data_alt_names = []
```

```python
if main_content:
    list_sections = main_content.find_all('div', {'class': 'ecs-list'})
    
    for list_section in list_sections:
        for section in list_section.find_all('div', {'class': 'section'}):
            for item in section.find_all('div', {'class': '_item'}):
                data_alt_name = item.get('data-alt-name')
                if data_alt_name:
                    data_alt_names.append(data_alt_name)

print(data_alt_names)
```
