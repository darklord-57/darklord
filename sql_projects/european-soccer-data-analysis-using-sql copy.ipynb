##  from the first two tables lets create a table
##  that shows the league name for each country 

```python
%%sql
select c.name as COUNTRY, l.name as LEAGUE
from Country c, League l
where c.id = l.id;
```

## FIND THE NUMBER OF MATCHES PLAYED BY EACH 
## COUNTRY 

```python
%%sql
SELECT c.name as Country, count(*) as '#matches'
FROM Country c, Match m
where c.id = m.country_id
group by c.name
order by count(m.country_id) desc;
```

### now lets move on to players 
### lets find the oldest 5 and youngest 5 players and see the variation in their overall ratings

```python
%%sql
select id,player_name,overall_rating,birthday
from 
(select * from (
select *
from Player p, Player_Attributes pa
where p.id = pa.id
order by birthday limit 5 ) as tab1
UNION 
select * from (
select *
from Player p, Player_Attributes pa
where p.id = pa.id
order by birthday desc limit 5) as tab2 )
order by birthday 
```

### lets find the total goals scored by each team

```python
%%sql
select team_long_name as '#Team', sum(home_team_goal) as '#total_goals_scored'
from Match m, Team t 
where t.team_api_id = m.home_team_api_id
group by team_long_name
order by sum(home_team_goal) desc limit 10
```

### find the number of games played by each team, to make it fair. 

```python
%%sql
SELECT  t.team_long_name AS Team, c.name AS Country, 
COUNT(m.home_team_api_id) AS no_of_matches,
SUM(m.home_team_goal) AS total_goals,
cast((SUM(m.home_team_goal)) as float) /COUNT(m.home_team_api_id) as ratio
FROM Team t, Match m, Country c
WHERE t.team_api_id = m.home_team_api_id AND m.country_id = c.id 
GROUP BY t.team_api_id
ORDER BY total_goals DESC
LIMIT 10;
```

```python
%%sql
select season, 
         CASE WHEN home_team_goal > away_team_goal THEN Team1.team_long_name
              WHEN home_team_goal < away_team_goal THEN Team2.team_long_name
                ELSE 'DRAW'
                END as team_won,
        CASE WHEN home_team_goal > away_team_goal THEN home_team_goal
              WHEN home_team_goal < away_team_goal THEN away_team_goal
                ELSE home_team_goal
                END as goals_scored,
        Team1.team_long_name as home_team,
        Team2.team_long_name as away_team
from Match 
join Team as Team1 on Match.home_team_api_id = Team1.team_api_id
join Team as Team2 on Match.away_team_api_id = Team2.team_api_id limit 10;
```

### lets look at the goals scored each season :
